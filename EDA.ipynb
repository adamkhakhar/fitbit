```python
from typing import List
import os
import pandas as pd
import numpy as np
import time
import seaborn as sns
import datetime

PATH_TO_DATA = "/Users/ak/Desktop/Code/fitbit/Fitabase Data 4.12.16-5.12.16"


def list_files(directory: str) -> List[str]:
    """
    Returns a list of all the files in the specified directory.

    Args:
        directory: A string representing the directory path.

    Returns:
        A list of strings representing the file names in the directory.
    """
    # Check if the specified directory exists
    if not os.path.isdir(directory):
        raise ValueError(f"The directory path '{directory}' does not exist.")

    # Get a list of all the files in the directory
    files = os.listdir(directory)

    # Filter out any directories and return only the file names
    return [
        directory + "/" + f for f in files if os.path.isfile(os.path.join(directory, f))
    ]


def print_dataframes(mapping):
    """
    Prints each name and DataFrame in the specified mapping in pretty format.

    Args:
        mapping: A dictionary mapping names to pandas DataFrames.
    """
    for name, df in mapping.items():
        print(f"DataFrame: {name}")
        print(
            "Total Rows",
            len(df),
            "Rows With Valid ID",
            len(df[pd.to_numeric(df["Id"], errors="coerce").notnull()]),
        )
        print("-" * 30)
        display(df)
        print("\n")
```

## Join Data into Single DataFrame

```python
def combine_dataframes(mapping):
    """
    Combine the dataframes into a single dataframe including all columns, removing duplicated columns.

    Args:
        mapping: A dictionary mapping names to pandas DataFrames.

    Returns:
        A single pandas DataFrame containing all columns from the input DataFrames.
    """
    # Create an empty list to store the individual dataframes
    dfs = []

    # Iterate over each key-value pair in the mapping dictionary
    for name, df in mapping.items():
        # Add the DataFrame to the list
        dfs.append(df)

    # Concatenate the individual dataframes into a single dataframe
    combined_df = pd.concat(dfs, axis=0, ignore_index=True, sort=False)

    return combined_df
```

```python
# parallelized implementation
def create_timestamp_column(
    df,
    timestamp_cols=[
        "ActivityDate",
        "ActivityDay",
        "Date",
        "ActivityHour",
        "ActivityMinute",
        "Time",
        "SleepDay",
    ],
):
    """
    Given a sparse DataFrame with multiple timestamp columns, creates a single timestamp column and removes the predecessors.

    Args:
        df: A pandas DataFrame with multiple timestamp columns.
        timestamp_cols: List of cols to get timestamp from

    Returns:
        A pandas DataFrame with a single timestamp column and the predecessors removed.
    """
    for col in timestamp_cols:
        s = time.time()
        if col not in ["ActivityDay", "ActivityDate"]:
            df[col] = pd.to_datetime(df[col])
        else:
            df[col] = pd.to_datetime(df[col], format="%m/%d/%Y")
        print("time", col, int(time.time() - s))

    s = time.time()
    timestamp_df = df[timestamp_cols]

    # Fill missing values using both backward fill and forward fill methods
    filled_df = timestamp_df.bfill(axis=1).ffill(axis=1)

    # Set the 'timestamp' column equal to the value of the first filled timestamp column
    df["timestamp"] = filled_df.iloc[:, 0]
    print("time", "timestamp", int(time.time() - s))

    # Remove the original timestamp columns
    df.drop(timestamp_cols, axis=1, inplace=True)

    return df


merged_df = create_timestamp_column(merged_df)
assert merged_df["timestamp"].isna().sum() == 0
merged_df
```

# Create Target and Features

### Target
For each day, for each user (ID), value of VeryActiveMinutes

```python
def extract_very_active_minutes(df):
    """
    Given a DataFrame, groups the data by ID and day, and selects the 'VeryActiveMinutes' column.
    Returns a DataFrame with columns 'timestamp', 'ID', and 'VeryActiveMinutes' and removes rows with non-numeric values in 'VeryActiveMinutes'.

    Args:
        df: A pandas DataFrame containing 'Id', 'timestamp', and 'VeryActiveMinutes' columns.

    Returns:
        A pandas DataFrame with columns 'timestamp', 'Id', and 'VeryActiveMinutes' without rows containing non-numeric 'VeryActiveMinutes'.
    """

    # Group by 'ID' and day, and select the 'VeryActiveMinutes' column
    df["VeryActiveMinutes"] = pd.to_numeric(df["VeryActiveMinutes"], errors="coerce")
    grouped_df = (
        df.groupby([df["Id"], df["timestamp"].dt.date])["VeryActiveMinutes"]
        .sum()
        .reset_index()
    )

    # Rename the 'TimeStamp' column back to its original name
    grouped_df.rename(columns={"timestamp": "Date"}, inplace=True)

    # Remove rows with non-numeric 'VeryActiveMinutes' values
    grouped_df.dropna(subset=["VeryActiveMinutes"], inplace=True)

    grouped_df.set_index(["Id", "Date"], inplace=True)

    return grouped_df
```

# Intensity and Steps Feature Creation

```python
def compute_minute_level_sum_to_hour(df, col_pref, new_col_name):
    cols = df.iloc[:, df.columns.str.startswith(col_pref)]
    sum_row = cols.sum(axis=1)
    df[new_col_name] = sum_row
    return df

compute_minute_level_sum_to_hour(index_df, "Intensity", "HourSumIntensity")
index_df[["HourSumIntensity", "timestamp"]]
```

```python
# Compute avg, sum of steps/intensities per hour per day

features_df["MaxHourMinIntensity"] = index_df.groupby(['Id', 'Date']).agg({'HourSumIntensity': 'max'})[["HourSumIntensity"]]
features_df["SumMinIntensity"] = index_df.groupby(['Id', 'Date']).agg({'HourSumIntensity': 'sum'})[["HourSumIntensity"]]
features_df["SumMinIntensityTop2Hr"] = index_df.groupby(['Id', 'Date'])[["timestamp", "HourSumIntensity"]].apply(lambda x: x.nlargest(2, 'HourSumIntensity').sum())["HourSumIntensity"]

features_df["MaxHourSteps"] = index_df.groupby(['Id', 'Date']).agg({'HourSumSteps': 'max'})[["HourSumSteps"]]
features_df["SumSteps"] = index_df.groupby(['Id', 'Date']).agg({'HourSumSteps': 'sum'})[["HourSumSteps"]]
features_df["SumStepsTop2Hr"] = index_df.groupby(['Id', 'Date'])[["timestamp", "HourSumSteps"]].apply(lambda x: x.nlargest(2, 'HourSumSteps').sum())["HourSumSteps"]
features_df
```
