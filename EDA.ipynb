```python
from typing import List
import os


def list_files(directory: str) -> List[str]:
    """
    Returns a list of all the files in the specified directory.

    Args:
        directory: A string representing the directory path.

    Returns:
        A list of strings representing the file names in the directory.
    """
    # Check if the specified directory exists
    if not os.path.isdir(directory):
        raise ValueError(f"The directory path '{directory}' does not exist.")

    # Get a list of all the files in the directory
    files = os.listdir(directory)

    # Filter out any directories and return only the file names
    return [
        directory + "/" + f for f in files if os.path.isfile(os.path.join(directory, f))
    ]
```
